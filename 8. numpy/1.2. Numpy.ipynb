#### __Difference numpy and python lists__
1. Memory usage
* Python list :- Stores elements as seperate python objects, each with extra metadata(type info, reference).

They are efficient but uses more memory.

* Nampy array:- Stores elements in a single continous blockof memory, all of the same type(int32, float64, etc)

They are more compact and efficient

Nampy are stored more efficiently than python lists.

Lists in python are essentially arrays of pointers to objects, with each pointer requiring additional space.

In contrast, numpy arrays directly store data in adjacent blocks of memory.

This compact and continous storage reduces the memory footprint and improves cache utilisation, which is critical for performance in large scale.

2. Speed(performance)
Numpy arrays are implemented in C and use vectorized operations.

This means they perform element wise without python loops.

One of the most powerful features of numpy arrays is there support for vector operations.

This means that operations on arrays, such as additions, multiplaction and complex mathematical operations, can be performed on all elements of the array simultaneously.

Numpy is often much faster - often 10 - 100x faster than lists

This is not only more syntactically convenient but also significantly faster than looping over elements as is common with python lists.

3. Data type consistency
* Python lists can hold multile data types eg ["Two", 5, False]
* Numpy array must have one data type.

This consistency allows numpy to perform mathematical operations efficiently.

Lists contain __heterogeneous data__ types(combination of str, int, even list), while numpy arrays can only store values of the same data type.

This restriction allows numpy to optimise storage and computation, as operations on homogeneously typed elements are mor efficient.

4. Mathematical operations
Python lists can't do element wise math directly

* * __[1,2,4] + [5,6,7]__  -- concantenates lists

Nampy arrays perform element-wise mathematical operations in our case additions
* * np.array([1,2,3]) + np.array([4,5,6]) = [5,7,9]

5. Built in functions
Numpy has tons of optimized mathematical functions eg, average, starndard deviation, sum.

Python lists would recuire you to loop manually.

#### __Dot product__
---
The dot produt (also know as scalar product or inner product) is a way of combining two vectors(or matrices) to reduce single number (scalar) or another matrix - depending on the shape.

Dot product is just another way of finding patterns between two sets of numbers

There are two forms of multiplacation between matrixes:

* * Element-wise multiplacation or hardamad product 
* * Dot product

There is a picture accompanied with this notebook explaining the difference in the two.

The difference is in how execution hapens

Mathematical, for two 1D vectors:

* * A=[a1, a2, a3]
* * B=[b1, b2, b3]

The dot product is:

* * A.B = a1b1 + a2b2 + a3b3

Example of dot product of 1D array(vectors):

````python
A = np.array([1,2,3])
B = np.array([4,5,6])

dot = np.dot(A,B)
print(dot)
````

The output is 32

This is what is happening behind the scenes.

* * (1*4) + (2*6) + (3*6) = 4 + 10 + 18 = 32

Example 2: Dot product of 2D arrays(matrices)

C = [1,2]

````python
np.random.seed(7)
matrix_1 = np.random.randint(10, size = (5,3))
print(matrix_1.shape)
matrix_2 = np.random.randint(10, size = (5,3))
print(matrix_2.shape)
````

````python
#Now let's calculate the dot product

dot_product = np.dot(matrix_1, matrix_2)
````

We are getting an error that : __ValueError: shapes (5,3) and (5,3) not aligned: 3 (dim 1) != 5 (dim 0)__

We are getting an error because there are rules when it comes to working with dot product.

There are compatibility rules!

The compatibility rules for dot product are:

* * 1. For two arrays to be compatible in a dot product, the last dimension of A must equal the second-to-last dimension of B.

Illustration:

* * A.shape = (m,n)

* * B.shape = (n,p)

The dot product is valid because:

* * __n==n(the inner dimensions match)__

__THE RESULT SHAPE BECOMES (M,P)__

* * 2. Simple case: __Vectors__

1D arrays(vectors)

* * A:(n,)
* * B:(n,)

Dot product is allowed because both have length n.

__Result: SCALAR__

another example - different shape

* * A:(n)

* * B:(p)

These two are not compatible because the shape differ - it will result to an error.

* * 3. __Matrix * Vector__

* * A: (m,n)
* * B: (n,)

The two are compatible because __n mathches n__

__THE RESULT SHAPE WILL BE (m,)__

another example:

* * A: (m,p)
* * B: (n,)

The two will not be compatible because __p is not equal to n__

* * __Matrix * Matrix__

* * A: (m,n)
* * B: (n,p)

inner dims: n vs n = valid/compatible

__RESULT SHAPE (m,p)__

another example-incompatible:

* * A: (c,d)
* * B: (k,l)

inner dims: d vs k = invalid/incompatible

__POINT TO NOTE__

* __THE RESULTS ASSUME THE SHAPE OF THE FIRST DIMENSION OF A__

Why:

Because matrix multiplacation logic says:

* The last dimension of A and the second last of B collapse n.
* The remaining dimension of A define the raw structure of the results.

Random numbers look random to us but what they really are is pseudo random numbers.

Meaning, whenever we run the codee without a seed, it sets the seed to be something random every single time we run the code but if we set our seed manualy, we are going get the same random numbers everytime we run the code.

We can see this by using __np.random.seed()__

Everytime we run our code without a random seed, it will keep on producing different elements in our array.

Let's say we were to share our code with someone else and we wanted them to run our code on there laptop and get the same results as us, without a random seed it would be impossible.

Random seed allows us to still generate random numbers, but ones which are reproducable.  

If we set our random seed to be let's say zero, we will still get random numbers but the same random numbers no matter how many times we run our code. - illustrations down below

It creates random numbers but ones which are aligned with the seed

We use random numbers when we want to share our code and we want the other person to produce the same random numbers as us thus producing the same results.

````python
#Sum
sum_= np.sum(array_3)
print(sum_)
print("-" * 50)

#mean
mean = np.mean(array_3)
print(mean)
print("-" * 50)

#standard diviation - the measure of spread within our data
#It is the measure of how spread out a group of numbers is from the mean
#It is the square root of the variance
std = np.std(array_3)
print(std)
print("-" * 50)

#variance
# Variance is the measure of the average degree to which each number is different to the mean
# Higher variance = wider range of numbers
# Lower variance = lower range of numbers

var = np.var(array_3)
print(var)
print("-" * 50)

#minimum value
min_ = np.min(array_3)
print(min_)
print("-" * 50)

#maximum value
max_ = np.max(array_3)
print(max_)
print("-" * 50)

#The position of the minimum value
index_min = np.argmin(array_3)
print(index_min)
print("-" * 50)

#The position of the maximum value
index_max = np.argmax(array_3)
print(index_max)
print("-" * 50)

# Performing aggregation on columns and rows using axis
# adding all the columns
sum_columns = np.sum(array_3, axis = 0)
print(sum_columns)
print("-" * 50)

# adding all the rows
sum_rows = np.sum(array_3, axis = 1)
print(sum_rows)
````

#### __Broadcasting__
---
Broadcasting allows numpy to perform operations on arrays with different shapes by virtually expanding dimensions so that they match the larger array.

With broadcasting, we can virtually there dimensions.

Numpy will virtually expand dimensions so that ht smaller array matches the larger array's shape.

Broadcasting is a set of rules that allows arrays of different shapes to be used together in arithmetic operations without manually reshaping them.

Numpy automatically "stretches" (broadcasts) samaller arrays so the operation can proceed

It doesn't create full copies; it creates virtual dimensions to make operations efficient.

There are a couple of rules when it comes to broadcasting:

Two arrays are compatible when:

* * They are equal
* * One of them is one

When comparing shapes, __we compare them from the rightmost dimension backwards__

When we say compare from the right, we mean:

__Take the rightmost dimension of the top array, compare it witj the rightmost dimension of the botom array.__

Let's go through some examples:

A â†’ shape **(4, 3)**

B â†’ shape **(3, 2)**

we want to check if A and B can broadcast.

__Step 1 - Align them from the right__

```
A:   (4, 3)
B:   (3, 2)
```
__Step 2 - Start comparing *from the rightmost numbers*__

__ðŸ”¹ Compare LAST dimension__

Top: **3**
Bottom: **2**

* Are they equal? â†’ âŒ

* Is one of them 1? â†’ âŒ

* âž¡ï¸ FAIL â†’ broadcasting impossible.

We **stop here** because one pair already failed.

---

__Another example__

* * A â†’ (5, 4, 3)

* * B â†’ (     3)

Align them from the right:

```
A: (5, 4, 3)
B:       (3)
```

* 3 vs 3 â†’ âœ” OK
* 4 vs (nothing) â†’ treat missing as 1 â†’ âœ” OK
* 5 vs (nothing) â†’ treat missing as 1 â†’ âœ” OK

Broadcasting works.

â­ So your rule is:

When comparing shapes:

ðŸ‘‰ **Look at the rightmost values first**

ðŸ‘‰ Compare **top with bottom**

ðŸ‘‰ Move left step by step

ðŸ‘‰ If any pair fails, broadcasting fails

__Broadcast is fundamentally a validity check:__

1. Start compare shapes from the rightmost dimension.

2. For each pair of dimension, check:

* * They are equal or 
* * One of them is 1

3. If a __**dimension pair**__ does not satisfy these rules, broadcasting immediately fails.

4. If all dimension pairs pass, broadcasting is possible.

We don't need to modify anything manually- Numpy just checks compatibility.

````python
np.random.seed(7)

broad_1 = np.random.randint(10, size = (5,6))
print(broad_1.shape)

broad_2 = np.random.randint(10, size=(1,6))
broad_2.shape
````

````python
#The two arrays can be broadcasted because one pair is equal and the other has a one
broad_1 * broad_2
````

#### __Filtering__
---
Filtering refers to the process of selecting elements from an array that match a given condition

__POINT TO NOTE!!!__

__Since numpy uses C style arrays, instead of the **'and'** or the **'or'** operator in python, we use **'&'** for **'and'** and **'|'** for **'or'**__

Le's do some examples:


````python
#Let's say we have two classes of students
ages = np.array([[21,17,19,20,16,30,18,65],
                 [39,22,15,99,18,19,20,21]])

#Teenagers - less than 18
teenagers = ages[ages < 18]
print(teenagers)

#Adults - between 18 and 65
adults = ages[(ages >= 18)& (ages < 65)]
print("-" * 50)
print(adults)

#Less than 18 or greater than 65

comparison = ages[(ages < 18) | (ages > 65)]
print("-" * 50)
print(comparison)

seniors = ages[ages >= 65]
print("-" * 50)
print(seniors)

even = ages[ages % 2 == 0]
print("-" * 50)
print(even)

odds = ages[ages % 2 != 0]
print("-" * 50)
print(odds)


#When filtering and creating a new array, if we need to preserve the original shape of our array we use the where function
#Let's say we want to filter out all the teens to remain with youths and seniors alone

adults_only = np.where(ages >= 18, ages, 0)
print("-" * 50)
print(adults_only)
#The where funtion will keep the original shape of our array but its alot slowere as compared to using boolean indexing
# We use where function if we need to ptrserve the orininal shape of our data or array/
#We also use it when we want to retain our original array
#Using boolean indexing flatens our data

#Or 
# This one modifies the original array - in place
ages[ages < 18] = 0
print("-" * 50)
print(ages)
````

#### __Sorting arrays__
---

Numpy has a function called __np.sort__ which we can use to sort arrays 

Its also has another functiuon called __np.argsort__ which is responsible for sorting the indexes in an array.

We can use this in the case where we want the index of the highest element in our array

We can also find the indexes of the highest and lowest arrays in our array:

* For 1D arrays :
* * np.argmin() - for min
* * np.argmax() - for max

* For 2D arrays:
* * np.argmax(array, axis = 0) - for max in columns
* * np.argmax(array, axis = 1) - for max in rows

* * np.argmin(array, axis = 0) - for min columns
* * np.argmin(array, axis = 1) - for min rows

__when looking for the index in a 2d array of the max or min value, the row axis is 1 - axis=1 and the column axis is 0, axis=0__

Examples:

````python
#let's create an array that we can work with
np.random.seed(7)
array_6 = np.random.randint(20, size = (3, 5))
print(array_6)

array_7 = np.random.randint(20, size = (1,5))
print(array_7)
````

````python
#Let's sort the array - array_6
np.sort(array_6)
````

````python
#Let's sort the indexes of our array
np.argsort(array_6)
````

````python
#Let's find the index of the minimum element in our 1D array 
print(array_6)
print(f"index_min; {np.argmin(array_7)}")
print("-" * 50)

# Let's find the index of the maximum elemnt in our 1D array
print(array_7)
print(f"index_max; {np.argmax(array_7)}")
print("-" * 50)


#Now let's find the indexes of the minimum elements in our 2D array, column wise
print(array_6)
print(f" indexes_min_columns; {np.argmin(array_6, axis = 0)}")
print("-" * 50)

#Let's find the indexes of the maximum elements in our 2 D array, row wise
print(array_6)
print(f"indexes_max_rows; {np.argmax(array_6, axis = 1)}")
````
