## Nested Dictionary

```python
alx_students = {
    "student_1":{
        "Name":"Kithusi Emmanuel",
        "Age":23,
        "Gender":"Male",
        "Program":"Data science"},
    "student_2":{
        "Name":"Jackline Njue",
        "Age":28,
        "Gender":"Female",
        "Program":"Data analytics"},
    "student_3":{
        "Name":"Mercy Mumo",
        "Age":25,
        "Gender":"Female",
        "Program":"Data Engineering"},
    "student_4":{
        "Name":"Esther Ngatia",
        "Age":21,
        "Gender":"Female",
        "Program":"Computer science"},
    "student_5":{
        "Name": "James Mwangi",
        "Age":26,
        "Gender":"Male",
        "Program":"Back end dev"}
}

alx_students
```

```python
# Modifying the name of student 1
alx_students["student_1"]["Name"]="Kithusi N Emmanuel"

alx_students
```

```python
# adding data in a nested dictionary

alx_students["student_6"]={
    "Name":"Ezekiel Chyalo",
    "Age":20,
    "Gender":"Male",
    "Program":"ICT",
}

alx_students
```

```python
# Accessing the data inside a nested dictionary

alx_students["student_1"]
```

```python
alx_students["student_1"]["Name"]
```

```python
alx_students["student_1"]["Name"]="Kithusi Emmanuel"

alx_students
```

```python
# The keys() method- this method returns or the keys in a dictionary

alx_students.keys()
```

```python
alx_students["student_1"].keys()
```

```python
# The values method - this method returns all the values in a dictionary

alx_students.values()
```

```python
alx_students["student_1"].values()
```

```python
# The items() method - this method returns both keys and there corresponding values

alx_students.items()
```
