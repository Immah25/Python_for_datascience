### Exercise 1

Given a list of deforested land in hectares over five years, calculate the total carbon emission over the years using built-in functions.

 Total carbon emissions are calculated by adding the hectares of deforested land and multiplying the total by a given carbon emissions factor of 40.

 Hint: 
 * Create a variable called `total_deforested_land`, the result of adding the hectares of the deforested land using the `sum` function.

```python
# my own version
deforested_land = [10, 15, 8, 12, 20]

carbon_emission_factor = 40

def total_carbon_emissions(deforested_land_over_the_years):
    """
    The fuction calculates the total carbon emission over the years by adding the hectares of deforested 
    land over the years and multiplying it by the given carebon emission factor

    parameters
    -----------
    The input is a list of hectares of deforested land over the years

    return
    -------
    The output is the total hectares of deforested land and the total carborn emission
    
    """
    total_deforested_land = 0
    for hec in deforested_land_over_the_years:
        total_deforested_land+=hec
    total_carbon_emissions = total_deforested_land * carbon_emission_factor
    return total_deforested_land, total_carbon_emissions
total_carbon_emissions(deforested_land)
```
