## Btwise operators
Bitwise operators are special symbols in Python that work on the binary (bit-level) representation of numbers.

They donâ€™t look at the number as a whole (like 5 or 7), but rather as bits (0s and 1s).

ðŸ§© The Bitwise Operators

| Operator    | Symbol | Meaning                                       | Example                                         | Result (Binary â†’ Decimal)  |    |       |                  |
| ----------- | ------ | --------------------------------------------- | ----------------------------------------------- | -------------------------- | -- | ----- | ---------------- |
| AND         | `&`    | Sets each bit to 1 **if both bits** are 1     | `a & b`                                         | `0101 & 0011 = 0001 â†’ 1`   |    |       |                  |
| OR          | `      | `                                             | Sets each bit to 1 **if at least one** bit is 1 | `a                         | b` | `0101 | 0011 = 0111 â†’ 7` |
| XOR         | `^`    | Sets each bit to 1 **if only one** bit is 1   | `a ^ b`                                         | `0101 ^ 0011 = 0110 â†’ 6`   |    |       |                  |
| NOT         | `~`    | Flips all bits (inverts 0 â†” 1)                | `~a`                                            | `~0101 â†’ -(0101 + 1) = -6` |    |       |                  |
| Left Shift  | `<<`   | Shifts bits left, adding 0s on the right      | `a << 1`                                        | `0101 â†’ 1010 â†’ 10`         |    |       |                  |
| Right Shift | `>>`   | Shifts bits right, dropping bits on the right | `a >> 1`                                        | `0101 â†’ 0010 â†’ 2`          |    |       |                  |


```python
a = 5  # 0101
b = 3  # 0011

print(a & b)   # AND â†’ 1
print(a | b)   # OR â†’ 7
print(a ^ b)   # XOR â†’ 6
print(~a)      # NOT â†’ -6
print(a << 1)  # Left shift â†’ 10
print(a >> 1)  # Right shift â†’ 2
```

## Identity operators

Identity operators are use to compare the memory locations of two objects

There two identity operators in python, 
>is

>is not

| Operator | Meaning                                                          |
| -------- | ---------------------------------------------------------------- |
| `is`     | Checks if **two variables refer to the same object in memory**   |
| `is not` | Checks if **two variables refer to different objects in memory** |


we can use identity operators (is and is not) with stringsâ€¦
but âš ï¸ you need to understand what they really do, because they donâ€™t always behave the way you expect.

Identity operators check memory location, not value equality.




```python
# ðŸ’¡ Example with strings

a = "hello"
b = "hello"

print(a == b)  # True â†’ because the *values* are the same
print(a is b)  # True (sometimes) â†’ because Python optimizes small strings

# ðŸ§©?Why a is b might be True

# Python interns (reuses memory for) short and simple strings like "hello".
# So both a and b might actually point to the same memory address â€” making a is b return True.
```

```python
# But this doesnâ€™t always happen.

# ðŸ§© Why a is b might be True

# Python interns (reuses memory for) short and simple strings like "hello".
# So both a and b might actually point to the same memory address â€” making a is b return True.

# But this doesnâ€™t always happen.

# âš ï¸ Example where it fails

a = "hello world!"
b = "hello world!"

print(a == b)  # âœ… True (values are equal)
print(a is b)  # âŒ Might be False (different memory locations)

# Python sometimes creates separate string objects in memory for longer or more complex strings, so is fails even though the text looks identical.
```

> For comparing string content its better to use the equality or the inequality operators

```python
#Identity operators with numbers
a = 10
b = 10.0

print("is:", a is b) #Output: False, as a and b are different types
```

```python
print("is not:", a is not b) #Output: True, as a and b are different types
```

```python
#Correct usage with equality
str1 = "Hello"
str2 = "Hello"

print("Correct Result with '==':", (str1==str2))
```
