### __inhertance__

__inherit. Extend. Override__

```python
#Parent classs
class Employee:
    def __init__(self, name, age, salary):
        self.name=name
        self.age=age
        self.salary=salary

    def work(self):
        print(f"{self.name} is working...")


#--------------------------------------------------------------------------
#Child class
class SoftwareEngineer_3(Employee):
    def __init__(self, name, age, salary, level):
        super().__init__(name, age, salary) # Extending our function
        self.level=level

    def work(self):
        super().work()
        print(f"{self.name} is coding....")

    def debug(self):
        print(f"{self.name} is debuging.....")


#------------------------------------------------------------------------
#Child classs 
class Designer(Employee):

    def work(self):         # Overiding the work method from the parent class
        print(f"{self.name} is designing a poster....")   

    def draw(self):
        print(f"{self.name} is drawing.....")


# polymophism -  it gives a way to use a class excatly like its parent but each child class keeps its methods as they are 

Employee_list=[SoftwareEngineer_3("James Mwangi", 26, 7000, "Junior"),
               SoftwareEngineer_3("Kithusi emmanuel", 26, 9000, "Senior"),
               Designer("Hadassah Esther", 24, 5000)]           


# for i in Employee_list:
#     i.work()
def motivate_employees(employees):
    for employee in employees:
        employee.work()



#-----------------------------------------------------------------------------------------------------------------------------------------------


print("-" * 50)
se = SoftwareEngineer_3("James Mwangi", 26, 7000, "Junior")
print(se.age, se.name, se.salary, se.level)
print("-" * 50)

d = Designer("Hadassah Esther", 24, 5000)
print(d.age, d.name, d.salary)
print("-" * 50)

se.work()
d.work()
print("-" * 50)

se.debug()
d.draw()
print("-" * 50)

motivate_employees(Employee_list)
print("-" * 50)
```

### __Encapsulation and Abstraction__

This is the mechanism of hiding of data implementation

This means instance variables are kept private

This instance variables can be accessed and modified only using functions called getter methods

We can also do this to methods, where methods can be kept private so that they should be used internally and not from the outside

```python
class SoftwareEngineer_4:

    def __init__(self, name, age):
        # public attributes
        self.name=name
        self.age=age 
        # private attribute- This is not completly private but we shouldn't or we are not supposed to access it from the outside
        # When we use two underscores it becomes completely private
        self._salary = None
        self._num_bugs_solved = 0  

    def code(self):
        self._num_bugs_solved +=1 

    
    # getter
    def get_salary(self):
        return self._salary
    
    #setter
    def set_salary(self, base_value):
        # check value, enforce constraints
        # if base_value < 1000:
        #     self._salary = 1000
        # if base_value > 20000:
        #     self._salary = 20000  
        self._salary= self._calculate_salary(base_value)
    
    def _calculate_salary(self, base_value):   # An example of Abstraction principle the complex stuff is done internally
        if self._num_bugs_solved < 10:
            return base_value
        if self._num_bugs_solved > 100:
            return base_value * 2
        return base_value * 3
        

#-------------------------------------------------------------------------------------------------------------------------------------------------
se = SoftwareEngineer_4("James Mwangi", 26)
print("-" * 50)
```

```python
for i in range(70):
    se.code()

se.set_salary(6000)
print(se.get_salary())
```

### __@properties @setters @deleters__

```python


class SoftwareEngineer:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self._salary = None
        self._num_bugs_solved = 0

    def code(self):
        self._num_bugs_solved += 1

    # property getter
    @property
    def salary(self):
        return self._salary

    # property setter
    @salary.setter
    def salary(self, base_value):
        self._salary = self._calculate_salary(base_value)

    # property deleter
    @salary.deleter
    def salary(self):
        print("Deleting salary...")
        self._salary = None

    # private helper method
    def _calculate_salary(self, base_value):
        if self._num_bugs_solved < 10:
            return base_value
        if self._num_bugs_solved > 10:
            return base_value * 2
        return base_value * 3
    
```

```python
se = SoftwareEngineer("Emmanuel", 25)
for i in range(70):
    se.code()


se.code()
se.code()
se.salary = 6000        # Calls @salary.setter
print(se.salary)        # Calls @property (getter)
del se.salary           # Calls @salary.deleter
```

### __Formal Abstraction using Abstract__

```python
from abc import ABC, abstractmethod
```

```python
class Animal(ABC):
    def __init__(self, name):
        self.name=name

    def walk(self):
        print(f"{self.name} is walking...")

class Talking(ABC):
    @abstractmethod
    def talk(self):
        ...


class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)

    def wag_tail(self):
        print(f"{self.name} is wagging its tail..")


class TalkingDod(Dog, Talking):
    def __init__(self, name):
        super().__init__(name)

    def talk(self):
        print(f"{self.name} is saying hello...")

```

```python
pluto = Dog("pluto")
print(pluto.name)
pluto.wag_tail()


rufus= TalkingDod("rufus")
rufus.talk()
rufus.wag_tail()
rufus.walk()
```

```python
# Abstract classes - A class that cannot be instantiated on its own; Meant to be subclassed
# They can contain abstract methods, which are declared but have no implementation
#         Abstract benefits:
#                 1. Prevents instantiation of the class itself
#                 2. Requires children to use inherited abstract methods


class Vehicle(ABC):
    @abstractmethod
    def go(self):
        ...

    @abstractmethod
    def stop(self):
        ...


class Car(Vehicle):
    def go(self):
        print("You drive this car")
    
    def stop(self):
        print("You stop this car")

class Mortocycle(Vehicle):
    def go(self):
        print("You ride the motocycle")

    def stop(self):
        print("You stop the mortocycle")



    
# When we say we have to implement, what de we mean
# with abstract methods, we declare them but we do not define them. We will define them within eac of the children classes That inherit from vehicle
#If a class is inheriting from a parent that is abstract and there is abstract methods, we have to finidh defining those methods
# When we include abstract methods within a parent, they act as sets of check and balances, all vehicles should be able to go and stop 
# like in our example. If we forget to define one of the methods, we will recieve a TypeError

car1 = Car()
```

```python

# Abstract Base Class
class Employee(ABC):
    def __init__(self, name, base_value):
        self.name = name
        self.base_value = base_value

    @abstractmethod
    def calculate_salary(self):
        pass


# Subclass 1
class Manager(Employee):
    def calculate_salary(self):
        return self.base_value * 3


# Subclass 2
class Developer(Employee):
    def calculate_salary(self):
        return self.base_value * 2


# Subclass 3
class Intern(Employee):
    def calculate_salary(self):
        return self.base_value


# Function using isinstance()
# def display_salary(employee):
#     if not isinstance(employee, Employee):
#         print("Error: Object must be an Employee instance.")
#         return

#     print(f"{employee.name}'s salary is {employee.calculate_salary()}")
def display_salary(employee):
    # if not isinstance(employee, Employee):
    
    #     print("Error: Object must be an Employee instance.")
    #     return
    # print(f"{employee.name}'s salary is {employee.calculate_salary()}")
    if isinstance(employee, Employee):
        return f"{employee.name}'s salary is {employee.calculate_salary()}"
    return ("Error: Object must be an Employee instance.")
    


# Creating objects
m = Manager("Alice", 5000)
d = Developer("Bob", 3000)
i = Intern("Charlie", 1500)

# Display salaries
print(display_salary(m))
print(display_salary(d))
print(display_salary(i))

# Example of invalid object
display_salary("Not an Employee")  # will trigger the isinstance() check

```
