#### __Static Methods__
---
A static method is a method inside a class that does not depend on any instance or class data.

It's just a Regular function placed inside a class for organization.

It:-

* Doesn't take self (so it doesn't know anything about the instance).
* Doesn't take class(so it doesn't know anything about the class itself.)
* It is used for utility/helper functions that are related to the class conceptually, but don't need access
to its attributes

Example:-

```python
class MathTools:

    @staticmethod
    def add(a, b):
        return a+b
    
MathTools.add(5, 6)

# Or

Math = MathTools()

Math.add(50, 60)
```

Here, @staticmethod is a decorator - it tells pyhton this method is static.

It doesn't use self, cls

You can call it:

> MathTools.add(5, 6)

or even:

Math=MathTools
Math.add(50, 60)

It works the same either way because it does not rely on instance data.

#### __Analogy__
Imagine we run a car factory again:

* We might have class methods like total_cars_built() information shared by the entire company factory.
* But sometimes you need a utility tool like calculate_tax(price) - which isn't tied to any specific car or
factrory itself.

That is our static method, a tool inside the class tool box, not belonging to any car.

Another example:-

```python
class Car7:
    def __init__(self, brand, price):
        self.brand=brand
        self.price=price

    @staticmethod
    def calculate_tax(price):
        return price * 0.16 #16% tax
    

Car7.calculate_tax(5000)
```

#### __Class methods__
This is a method that operates on itself not on individual instances. 

It is used when we want to work with the class level data - data shared across all instances.

To create one, we use the @classmethod decorator.

Instead of self, it takes a special first parameter called cls, which refers to the class itself.

Example:-

```python
class Car8:

    total_cars=0 #Class attribute

    def __init__(self, brand):
        self.brand=brand
        Car_8.total_cars+=1

    @classmethod
    def show_total_cars(cls):
        print(f"The total built cars are: {cls.total_cars}")

car1 = Car_8("Mascedes")
car2 = Car_8("Mazda")
car3 = Car_8("BMW")

Car_8.show_total_cars()
```

#### __Inheritance__

This is the process by which one class take takes on the attributes and methods of another class.

The newly formed class is called a child class and the other is called a parent class.

Inheritance means a new class (child class) can reuse the attributes and methods of an existing class (parent class) - without rewriting them.

#### __Analogy__
Think of inheritance like a family:

* The parent class = Parent
* The child class = Child

A child inherits traits from there parents(like Hair, color, etc), but they can also have there unique traits.

Exanple:

```python
#Parent class
class Car10:
    def __init__(self, brand, year):
        self.brand=brand
        self.year=year

    def start_engine(self):
        print(f"{self.brand} engine started!")


#Child class
class ElectricCars(Car10):
    def __init__(self, brand, year, battery_capacity):
        #Inherited attributes from the Car10 class
        super().__init__(brand, year)
        #New attributes for the electric car
        self.battery_capacity=battery_capacity

    # instance method
    def charge(self):
        print(f"{self.brand} is charging with {self.battery_capacity} KWH battery")

car1 = ElectricCars("Tesla", 2025, 90)
print(car1.brand, car1.year, car1.battery_capacity)
```

#### __Explanation__
* Class ElectricCars(Car_10) = means ElectricCars inherits everything from Car_10
* super().\_\_init__ (brand, color) - calls the parent class constructor to set those attributes.
* The child class can now:

    * Use all parent methods(start_engine)

    * Add new ones(charge)

    * Even overriding existing ones, (eg. redefine start_engine to behave differently)

#### __super() - calling the parent__

super() let's us access methods and attributes from the parent class especially useful when the child class wants to extend (not to completely replace) what the parent already does

#### __Analogy__
Imagine your dad runs a restuarant and you the child manage it

You still call your dad when you need his original reciepe, but you can add your own twist before serving

Example: 

#### __Extending the parent behavior:__

```python
class Car11:
    def __init__(self, brand, year):
        self.brand=brand
        self.year=year

    def engine_start(self):
        print(f"{self.brand} engine started!")

    
class ElectricCars1(Car11):
    def __init__(self, brand, year, battery_percentage):
        super().__init__(brand, year)
        self.battery_percentage=battery_percentage

    def engine_start(self):
        #Extending parent behavior
        super().engine_start()
        print("Slently running on electric mode")

car1 = ElectricCars1("Tesla", 2025, 90)

car1.engine_start()
```

#### __Overriding or replacing the parent behavior__
When a child class defines a method with the same name as the parent's, the child's version takes over - it overrides the parent method.

#### __Analogy__
You inherit your dad's cooking skills, but you prefer to cook differently - your way replaces his reciepe!

Example:

```python
class Car12:
    def fuel_type(self):
        print("Runs on petro")

class ElectricCars2(Car12):
    def fuel_type(self):
        print("Runs on electric")

Car12().fuel_type()
ElectricCars2().fuel_type()
```

#### __Polymorphism__
---
Polymophism literally means "Many forms"

In OOP, it means, the same method name can behave differently depending on which object(class instance) calls it.

Simple analogy

Think of the word "play":

* A child plays with toys.
* A mucisian plays guitar.
* A footballer plays football.

same word - "Play", different behaviors depending on who is playing 

That is polymorphism.

Example in python - different classes, same method name:

```python
class Car13:
    def start_engine(self):
        print("Car engine start with a roar")

class ElectricCar:
    def start_engine(self):
        print("Electric car stating silently")

class Bike:
    def start_engine(self):
        print("Bike engine starting .....brmmm")


vehicals = [Car13(), ElectricCar(), Bike()]

for v in vehicals:
    v.start_engine()
```

What happened?

We called the same method name(start_engine) on each object.

Each class had its own version of that method.

Python automatically call the correct one based on the object type.

That is runtime polymorphism(behavior decide at runtime)

#### __Polymorphism with inheritance__
When you use inheritance, you can overide a method in the child class and still call it through a common interface.

Example:

```python
class Vehicle:
    def move(self):
        print("Vehicle is moving...")

class Car14(Vehicle):
    def move(self):
        print("Car is driving")

class Plane(Vehicle):
    def move(self):
        print("Plane is flying")

for transport in [Vehicle(), Car14(), Plane()]:
    transport.move()
```

The move() acts differently based on which class it belongs to.

##### __Key Benefits of polymorphism__

```python
def start_any_vehicle(means):
    means.start_engine()

start_any_vehicle(Bike())
start_any_vehicle(ElectricCar())
start_any_vehicle(Car13())
```

No matter the vehicle type, it just works. 

That's the power of polymorphism.

##### __In summary__

* Polymorphism = Same method name, different behaviors. eg start_engine(), behaves differently in car, electric car and bike
* It is achived by method overriding or sharing method names eg, child overrides parent method.
* The benefits are code flexibility and reusability eg same interface for different object types.

#### __Encapsulation__
---
Encapsulation means wrapping data (attributes) and behavior methods together inside a class and restricting direct access to some of that data.

It is the mechanism of hiding data implementation.

This means that instance variables are kept private and there is only one accesser method from the outside with which we can access or change these instance variables.

In other words, we decide what can be seen and what stays private inside our object.

#### __Analogy__

Think of a capsule medicine, the ingridience are hidden inside for protection.

You don't take the cupsule and take the raw powder, you just swallow it.

similar in oop:

* You interact with the class interface(it's methods)
* But you don't directly mess with its internal data.

In python, encapsulation is done using naming conventions:

* Public_var = __public__ = access anywhere
* _Protected_var = __protected__ = Should be treated as internal(convetion only)
* \_\_Private_var = __private__ = Hidden from outside access, errors when we try to access it.

Example:

```python
class Car15:
    def __init__(self, brand, speed):
        self.brand=brand  #Public Attribute
        self._engine_status="OFF" #Protected Attribute
        self.__speed=speed #Private attribute

    def start_engine(self):
        self._engine_status="ON"
        print(f"{self.brand} engine started.")

    def accelerate(self, increase):
        if self._engine_status == "OFF":
            print("Engine is off. please start the car first.")
        else: 
            self.__speed+=increase
            print(f"{self.brand} is now running at {self.__speed} km/h")

    def get_speed(self):
        return self.__speed
    
car1 = Car15("Mascedes", 50)

#If we run the acceleration function without starting the engine first,
#We get the following warning "Engine is off. please start the car first."
# car1.accelerate(50)

car1.start_engine()
car1.accelerate(50)
car1.get_speed()
```

```python
print(car1.get_speed()) #safe way to access private data. instead of car1.__speed
print(car1.brand)
print(car1._engine_status) #posible but not recommended
# print(car1.__speed) #Error, private attribute
```

What's actually happening.

Python doesn't have strict private variables like jave or c++, but uses name mangling for double underscores.

car1._engine_start = we still access this but it is not recomended

That is python's way of saying: 

"You can see this if you want to but you shouldn't touch it"

car.__speed = this one brings errors when we try to access it.

#### __Abstraction__
---
Abstraction means only showing the essential features of something and hiding the complex details underneath

Its like using a car - you can drive, brake and accelerate but you don't need to know how the engine works, or how fuel turns into motion.

In python terms its about creating a new interface and hiding the messy internal logic from the user.

So instead of this:-

se_calculate_salary(5000)

we can offer a simple method like:-

se.get_salary()

That hides all the inner math, checks and logic 

That is abstraction; The outside world interacts only with what it needs to know

#### __How python supports Abstraction__
Python gives us two main ways to apply abstraction:

1. By convention - using naming and design.

We hide the details by; 

* Prefixing variables/methods with '_' (eg. _calculate_salary).

* Exposing only high-level public methods (eg set_salary(), get_salary())

This is called informal abstraction - it is based on trust and discipline.

2. Formally using abstract base class(ABC)

This is where abstraction becomes explicit.

We use ABC module to create a base class __what must be done__ but not __how__.

#### __Summary__

* Abstraction means hiding complex details, showing simple interface. eg get_salary hides _calculate_salary.
* Informal abstraction means using '_' and design to hide internal data eg _private_method().
* Formal abstraction means using ABC and @abstractionmethod eg class Employee(ABC)

The goal is to make code easy to use and extend eg cleaner APIs and subclasses

#### __@setter__
---
The @age.setter decorator allows you to control how a varible is changed

It ensures:

* You can validate input 
* You can protect the variable
* You can add rules when setting a value

Example with setter:

```python
class Person:
    def __init__(self, age):
        self._age=age

    @property #This is our getter
    def age(self):
        return self._age
    
    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age=value


kithusi = Person(20)
kithusi.age=24      #setter runs
kithusi.age         #getter runs

#If we were to enter a negative value in our setter, then we would gst an error
# "Age cannot be negative"
```

#### __What a setter does and doesn't__
__It:__

* Updates an exiting attribute.
* Adds rules or validation before updating
* Protects the attributes from bad values
* Keeps our objects consistent

__It does not:__

* Create a new object
* Create a new variable
* Replace the whole instance

It only works on the same object already created using \_\_init__.


Putting everything together in an example:

```python
class BankAccount:
    def __init__(self, balance):
        self._balance = balance #protected

    @property       #GETTER
    def balance(self):
        return self._balance
    
    @balance.setter
    def balance(self, amount):
        if amount < 0:
            raise ValueError("Balance cannot be negative")
        self._balance += amount


acc = BankAccount(5000)

acc.balance             #getter
acc.balance = 10000     #Setter
acc.balance
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        #Encapsulation: protect the internal variable
        self._owner=owner
        self._balance=balance

    #---------------------------------------------
    #GETTER: Read the balance
    #---------------------------------------------

    @property
    def balance(self):
        return self._balance
    
    #---------------------------------------------
    #Setter: Control how balance updates are made
    #---------------------------------------------

    @balance.setter
    def balance(self, amount):
        if amount < 0:
            raise ValueError("Balance cannot be negative")
        self._balance=amount

    #-----------------------------------------------
    #HIGH-LEVEL METHODS (Abstraction)
    #-----------------------------------------------
    
    def deposit(self, amount):
        if amount < 0:
            raise ValueError("Deposit must be positive")
        
        #internal updates balance(setter handles rules)
        #self.balance is a funnction but we have called it like an instance due to the getter- @property property we set on it 
        self.balance=self._balance+amount
        print(f"Deposited {amount}. New balance is {self.balance}")

    def withdraw(self, amount):
        if amount < 0:
            raise ValueError("Withdrawal must be posive")
        
        if amount > self._balance:
            raise ValueError("Insufficient funds")
        
        #internal update balance
        self.balance = self._balance - amount
        print(f"Withdrawal {amount}. New balance is {self.balance}")
```

#### __Abstract Base Class - ABC__
---
ABC is a blueprint for other classes.

We use it to define methods that must be implemeted in all child classes, but not provide there implementation in the base class.

In pyhton, we create ABCs using the abc module, example:

```python
from abc import ABC, abstractmethod
```

```python
class Employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass # sub/child classes must implement this

class Manager(Employee):
    @property
    def calculate_salary(self):
        return 5000
    
class Developer(Employee):
    @property
    def calculate_salary(self):
        return 8000
    
#When we try to run Emmanuel = Employee(), we get an error, type error to be precise, that we can instantiate abstract class
# Emmanuel = Employee()

Emmanuel = Developer()
Emmanuel.calculate_salary
```

#### __Key points__
* You can't instantiate an abstract class directly.
* Any subclass must implement all abstract methods, or it becomes abstract to - meaning it cannot be instantiated.
* Helps ensure a consistent structure across related classes.

#### __isinstance() built in method__

__isinstance(object, classinfo)__ checks if an object is an instance of a class or subclass.

```python
class Animal:
    pass

class Dog(Animal):
    pass

dog = Dog()

print(isinstance(dog, Dog))         #True
print(isinstance(dog, Animal))      #True (Dog is a subclass of animal)
print(isinstance(dog, object))      #True, everything inherits from the object
```
